```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Importing the CSV file
```

```python
path = "rainfall.csv"
df = pd.read_csv(path)
```

```python
df.head(5)
```

```python
'''The heat map shows that the amount of rainfall in months of July to September is highest
   while also if the rains in October, November and December are high then the annual rainfall
   also increases giving good rain patterns'''
plt.figure(figsize=(10,5))
sns.heatmap(df[['JAN','FEB','MAR','APR','MAY','JUN','JUL','AUG','SEP','OCT','NOV','DEC','ANNUAL']].corr(),annot=True)
plt.show()
```

```python
#Make similar clusters of all the patterns observed
corr= df.corr()
sns.clustermap(corr)
```
